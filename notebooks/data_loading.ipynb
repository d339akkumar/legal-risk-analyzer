### Import Libraries 

```python
import json
import pandas as pd
from tqdm import tqdm
import os
```

### Load CUAD JSON

```python
DATA_PATH = "../data/raw/CUAD_v1.json"

with open(DATA_PATH, "r", encoding="utf-8") as f:
    cuad_data = json.load(f)

print("Total documents:", len(cuad_data["data"]))
```

### Convert CUAD → Clause Classification Dataset

```python
records = []

for doc in tqdm(cuad_data["data"]):
    for para in doc["paragraphs"]:
        for qa in para["qas"]:
            answers = qa["answers"]

            if answers:
                clause_text = answers[0]["text"]
                clause_type = qa["question"]

                if len(clause_text) > 30:   # remove tiny junk fragments
                    records.append({
                        "clause_text": clause_text.strip(),
                        "clause_type": clause_type.strip()
                    })

df = pd.DataFrame(records)

print("Total clauses extracted:", len(df))
df.head()
```

### Save Clean Dataset

```python
OUTPUT_PATH = "../data/processed/cuad_clauses.csv"
df.to_csv(OUTPUT_PATH, index=False)

print("Saved to:", OUTPUT_PATH)
```
