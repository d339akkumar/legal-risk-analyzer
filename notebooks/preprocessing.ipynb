## Preprocessing 

### Import Libraries 

```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
```

### Load Dataset

```python
df = pd.read_csv("../data/processed/cuad_risk_dataset.csv")

df.head()
```

### Text Clenaing

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"\s+", " ", text)      # normalize spaces
    text = re.sub(r"[^\w\s.,;:()%-]", "", text)  # keep legal punctuation
    return text.strip()

df["clean_text"] = df["clause_text"].apply(clean_text)

df[["clause_text", "clean_text"]].head()
```

### Encoding labels numerically 

```python
label_map = {
    "Low": 0,
    "Medium": 1,
    "High": 2
}

df["label"] = df["risk_level"].map(label_map)

df[["risk_level", "label"]].head()
```

```python
df = df.dropna(subset=["label"])

print("Remaining samples:", len(df))
print(df["risk_level"].value_counts())
```

### Train/Validation split

```python
X_train, X_val, y_train, y_val = train_test_split(
    df["clean_text"],
    df["label"],
    test_size=0.2,
    stratify=df["label"],
    random_state=42
)


print("Train size:", len(X_train))
print("Validation size:", len(X_val))
```
