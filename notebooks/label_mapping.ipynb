### Load Processed Clause CSV file

```python
import pandas as pd
import re

df = pd.read_csv("../data/processed/cuad_clauses.csv")
df.head()
```

### Extracting clean clause names

```python
def extract_clause_name(text):
    match = re.search(r'"(.*?)"', text)
    return match.group(1) if match else text

df["clean_clause"] = df["clause_type"].apply(extract_clause_name)

df[["clause_type", "clean_clause"]].head()
```

```python
df["clean_clause"].value_counts().head(41)
```

### Risk Mapping 

```python
risk_map = {

    # 🔴 HIGH RISK — serious legal/financial exposure
    "Cap On Liability": "High",
    "Uncapped Liability": "High",
    "Termination For Convenience": "High",
    "Exclusivity": "High",
    "Non-Compete": "High",
    "Indemnification": "High",
    "Liquidated Damages": "High",
    "Covenant Not To Sue": "High",
    "Most Favored Nation": "High",
    "Ip Ownership Assignment": "High",
    "Joint Ip Ownership": "High",
    "Change Of Control": "High",
    "Competitive Restriction Exception": "High",
    "Volume Restriction": "High",

    # 🟡 MEDIUM RISK — cost, compliance, operations
    "Payment Terms": "Medium",
    "Revenue/Profit Sharing": "Medium",
    "Minimum Commitment": "Medium",
    "Audit Rights": "Medium",
    "Post-Termination Services": "Medium",
    "Renewal Term": "Medium",
    "Insurance": "Medium",
    "Warranty Duration": "Medium",
    "Source Code Escrow": "Medium",
    "Price Restrictions": "Medium",
    "No-Solicit Of Employees": "Medium",
    "No-Solicit Of Customers": "Medium",
    "Rofr/Rofo/Rofn": "Medium",
    "Irrevocable Or Perpetual License": "Medium",
    "Unlimited/All-You-Can-Eat-License": "Medium",
    "Affiliate License-Licensee": "Medium",
    "Affiliate License-Licensor": "Medium",

    # 🟢 LOW RISK — informational / structural
    "Governing Law": "Low",
    "Expiration Date": "Low",
    "Effective Date": "Low",
    "Agreement Date": "Low",
    "Notice Period To Terminate Renewal": "Low",
    "Parties": "Low",
    "Document Name": "Low",
    "License Grant": "Low",
    "Anti-Assignment": "Low",
    "Non-Transferable License": "Low",
    "Third Party Beneficiary": "Low"
}
```

### Apply Risk Labels

```python
df["risk_level"] = df["clean_clause"].map(risk_map)

df[["clean_clause", "risk_level"]].head(10)
```

### Save  dataset

```python
df_final = df[["clause_text", "clean_clause", "risk_level"]]

df_final.to_csv("../data/processed/cuad_risk_dataset.csv", index=False)

df_final.head()
```
